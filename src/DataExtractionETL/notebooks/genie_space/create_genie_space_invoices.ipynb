# Create a Genie space for invoice extracted data

This notebook creates a [Genie space](https://docs.databricks.com/en/genie/) that answers questions about **invoice extracted data** using the `app_invoices_processed` table.

**API reference:** [Create Genie Space](https://docs.databricks.com/api/workspace/genie/createspace)

After the space is created, open it in the workspace and add the `app_invoices_processed` table (Configure Data) if needed.

```python
# %pip install databricks-sdk -q

from databricks.sdk import WorkspaceClient
from manage_genie_space import *
w = WorkspaceClient()

# Configuration – invoices
TABLE_FULL_NAME = "data_extraction.nikk.app_invoices_processed"
WAREHOUSE_ID = "4b9b953939869799"

GENIE_SPACE_TITLE = "Invoice extraction results"
DESCRIPTION = f"Answers about invoice extracted data using the {TABLE_FULL_NAME} table."

SERIALIZED_SPACE = f"""{{
  "version": 2,
  "data_sources": {{
    "tables": [
      {{
        "identifier": "{TABLE_FULL_NAME}",
        "column_configs": [
          {{"column_name": "file_name", "enable_format_assistance": true, "enable_entity_matching": true}},
          {{"column_name": "invoice_date", "enable_format_assistance": true}},
          {{"column_name": "invoice_sum", "enable_format_assistance": true}},
          {{"column_name": "seller", "enable_format_assistance": true, "enable_entity_matching": true}}
        ]
      }}
    ]
  }}
}}"""
```

```python
genie_space = get_existing_genie_space(w, GENIE_SPACE_TITLE)
```

```python
if not genie_space:
    response = create_genie_space(
        w, WAREHOUSE_ID, SERIALIZED_SPACE, GENIE_SPACE_TITLE, DESCRIPTION
    )
    print_space_info(response, created=True)
else:
    print_space_info(genie_space)
```
