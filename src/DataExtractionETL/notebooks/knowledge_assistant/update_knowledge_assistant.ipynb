# Update Knowledge Assistant – sync knowledge sources

This notebook runs **sync all Knowledge Sources** for a Knowledge Assistant (e.g. after new documents are added to the volume).

Used by the extraction ETL job after the pipeline run. Set `assistant_id` via job parameter or below when running interactively.

```python
from databricks.sdk import WorkspaceClient
from manage_knowledge_assistant import sync_knowledge_sources

w = WorkspaceClient()

# Assistant ID: from job parameter (assistant_id) or set when running interactively
try:
    assistant_id = dbutils.widgets.get("assistant_id")
except Exception:
    assistant_id = ""  # Set to your Knowledge Assistant id when running interactively

if not assistant_id:
    raise ValueError("assistant_id is required. Set it in the job parameters or in this notebook when running interactively.")
```

```python
sync_knowledge_sources(w, assistant_id)
print("Knowledge sources sync completed.")
```
